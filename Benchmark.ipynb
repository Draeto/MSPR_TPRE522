```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from skimage.feature import hog
import cv2  
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from skimage.transform import resize
import skimage.io as io
from tqdm import tqdm

IMG_PATH = "DATA"
Data = pd.DataFrame()
UPLOAD_PATH = "Upload"
folders = os.listdir(IMG_PATH)
```

```python
train_data = pd.DataFrame()


def load_data_images(folder_path):
    data = pd.DataFrame({"Path":[],"Target":[],"Features":[]})
    target = 0
    img_error = []
    for sub_folder in tqdm(os.listdir(folder_path)):
        sub_folder_path = os.path.join(folder_path,sub_folder)
        target +=1 
        for filename in os.listdir(sub_folder_path):
            img_dict = {}
            img_dict["Path"] = []
            img_dict["Target"] = []
            img_dict["Features"] = []
            img_path = os.path.join(sub_folder_path, filename)
            try :
                # print(img_path_list)
                img = io.imread(img_path, as_gray=True)
                resized_img = resize(img, (128, 128),anti_aliasing=True)  # Resize images to a common size
                fd = hog(
                    resized_img,
                    orientations=8,
                    pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1)
                )
                img_dict["Path"].append(img_path)
                img_dict["Target"].append(target)
                img_dict["Features"].append(fd)
                dict_df = pd.DataFrame(img_dict)
                # print(images_list)
            except :
                img_error.append(f"{filename}, {sub_folder}")
        
        data = pd.concat([data,dict_df])
        print(data)
    print("Image Error :\n" ,img_error)
    data.to_csv("data.csv",sep=";",index=False)
        
    return data 

data = load_data_images(IMG_PATH)

```
